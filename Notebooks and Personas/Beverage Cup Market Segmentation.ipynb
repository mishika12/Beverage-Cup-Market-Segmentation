# Beverage Cup Market Segmentation


### Part (A) Product affinity based segmentation

```python
import numpy as np 
import pandas as pd 
pd.options.display.max_columns = None

```

```python
df = pd.read_excel('mugs-data-for-segmentation-exercise.xlsx')
df
```

Section 1: Computing the characteristics of the affinity based segment for Product 3 

```python
attr_list = df.columns[1:-3] # choosing the required attributes/ descriptors 
attr_list
```

We know that the we need to compute the weighted avergaes for all the attributes/ descriptors where weight is equal to the probabilities. We will first focus on our candidate product that is Product 3 

```python
# Product 3
wavg_P3 = pd.DataFrame(columns=['Attribute', 'Weighted_Average_P3'])

for column in attr_list:
    weighted_avg = sum(df[column] * df['P3']) / sum(df['P3'])
    wavg_P3 = pd.concat([wavg_P3, pd.DataFrame({'Attribute': [column], 'Weighted_Average_P3': [weighted_avg]})], ignore_index=True)

wavg_P3
```

Section 2: Computing the characteristics of the affinity based segment for Product 1 and 2, overall mean and log-lifts of each descriptor 

```python
# Product 1 

wavg_P1 = pd.DataFrame(columns=['Attribute', 'Weighted_Average_P1'])

for column in attr_list:
    weighted_avg = sum(df[column] * df['P1']) / sum(df['P1'])
    wavg_P1 = pd.concat([wavg_P1, pd.DataFrame({'Attribute': [column], 'Weighted_Average_P1': [weighted_avg]})], ignore_index=True)

wavg_P1
```

```python
# Product 2

wavg_P2 = pd.DataFrame(columns=['Attribute', 'Weighted_Average_P2'])

for column in attr_list:
    weighted_avg = sum(df[column] * df['P2']) / sum(df['P2'])
    wavg_P2 = pd.concat([wavg_P2, pd.DataFrame({'Attribute': [column], 'Weighted_Average_P2': [weighted_avg]})], ignore_index=True)

wavg_P2
```

```python
#Overall mean of all descriptors 
overall_avg = pd.DataFrame(columns=['Attribute', 'Overall_Average'])

for column in attr_list:
    over_avg= np.average(df[column])
    overall_avg= pd.concat([overall_avg, pd.DataFrame({'Attribute': [column], 'Overall_Average': [over_avg]})], ignore_index=True)

overall_avg
```

```python
# Combining all the weighted averages and overall averages 

df_new = pd.concat([wavg_P1, wavg_P2, wavg_P3, overall_avg], axis= 1)
df_new = df_new.T.drop_duplicates().T
```

```python
# Computing lifts

df_new['Lift_P1'] = df_new['Weighted_Average_P1']/ df_new['Overall_Average']
df_new['Lift_P2'] = df_new['Weighted_Average_P2']/ df_new['Overall_Average']
df_new['Lift_P3'] = df_new['Weighted_Average_P3']/ df_new['Overall_Average']
df_new
```

```python
# Computing log-lifts

df_new['Log-lift_P1'] = df_new['Lift_P1'].apply(lambda x: round(pd.np.log10(x), 5))
df_new['Log-lift_P2'] = df_new['Lift_P2'].apply(lambda x: round(pd.np.log10(x), 5))
df_new['Log-lift_P3'] = df_new['Lift_P3'].apply(lambda x: round(pd.np.log10(x), 5))
df_new
```

```python
df_1 = df_new[['Attribute', 'Log-lift_P1',	'Log-lift_P2',	'Log-lift_P3']]
```

```python

import seaborn as sns
cm = sns.diverging_palette(10, 150, as_cmap=True)
df_1.style.background_gradient(cmap=cm).set_precision(5)
```

### Part (B) Classical segmentation

```python
df_s = df.iloc[:,1:-7] # segmentation basis variables
```

```python
# Using KMeans 
from sklearn.cluster import KMeans
import seaborn as sns

# for ensuring same answers
np.random.seed(410014)

# X is an np.array containing the data
X = np.array(df_s) 

ave_within_cluster_mean_sum_of_squares = {}
# for all k from 2 to 10
for k in range(2,11):
    kmeansModel = KMeans(n_clusters=k, n_init=50, max_iter=100)
    kmeansModel.fit(X)
    ave_within_cluster_mean_sum_of_squares[k] = (kmeansModel.inertia_) / X.shape[0]


ave_within_cluster_mean_sum_of_squares

    
```

```python
df_k = pd.DataFrame(data = list(ave_within_cluster_mean_sum_of_squares.items()), columns = ['K', 'MSS'])
df_k
```

```python
import matplotlib.pyplot as plt
plt.plot(df_k['K'], df_k['MSS'], marker='o', linestyle='-')
plt.xlabel('K')
plt.ylabel('MSS')
plt.title('Average within-cluster variance for all k')
plt.show()
```

Choosing optimal value of k=5 (Rationale given in PDF)


Section 1: Cluster Analytis and Profiling Analysis


```python
# Cluster Analysis for all segmentation basis variables when k = 5

np.random.seed(410014)
k = 5
# X is an np.array containing the data
X = np.array(df_s) 


kmeansModel = KMeans(n_clusters=k, n_init=50, max_iter=100)
kmeansModel.fit(X)


#ectracting the cluster labels and centroids
label, centroids= kmeansModel.labels_,kmeansModel.cluster_centers_

```

```python
# plotting the clusters for profiling set varibales (age and income)

# age and income columns
age = df['age']
income = df['income']


plt.figure(figsize=(10, 6))
label_map = {label: f'Cluster {label}' for label in set(label)}

# Plotting the data points with cluster labels
for l in set(label):
    indices = [i for i, x in enumerate(label) if x == l]
    plt.scatter(age.iloc[indices], income.iloc[indices], label=label_map[l])

plt.xlabel('Age')
plt.ylabel('Income')
plt.title('K-means Clustering: Age vs. Income')

plt.legend()
plt.show()
```

Using definition of Log-lifts in case of classic segementation = Log((unweighted average of "x" across all persons within segment "s")/(unweighted average of x across all persons regardless of segment)).

```python
# lIFTS

full_data['Lift_P1'] = full_data['Segment_1_Avg']/ full_data['Overall Avg']
full_data['Lift_P2'] = full_data['Segment_2_Avg']/ full_data['Overall Avg']
full_data['Lift_P3'] = full_data['Segment_3_Avg']/ full_data['Overall Avg']
full_data['Lift_P4'] = full_data['Segment_4_Avg']/ full_data['Overall Avg']
full_data['Lift_P5'] = full_data['Segment_5_Avg']/ full_data['Overall Avg']
full_data
```

```python
#Log-lifts

full_data['Log-lift_P1'] = full_data['Lift_P1'].apply(lambda x: round(pd.np.log10(x), 5))
full_data['Log-lift_P2'] = full_data['Lift_P2'].apply(lambda x: round(pd.np.log10(x), 5))
full_data['Log-lift_P3'] = full_data['Lift_P3'].apply(lambda x: round(pd.np.log10(x), 5))
full_data['Log-lift_P4'] = full_data['Lift_P4'].apply(lambda x: round(pd.np.log10(x), 5))
full_data['Log-lift_P5'] = full_data['Lift_P5'].apply(lambda x: round(pd.np.log10(x), 5))


full_data

```
